```python
# Assigning the dataset,EmergencyTriageDataset_Reduced_Dirty.csv, to a variable
FILE_PATH= '/content/drive/MyDrive/Colab Notebooks/CariSurg_Triage_Test/EmergencyTriageDataset_Reduced_Dirty.csv'

# Reading the file dataset
raw_data = pd.read_csv(FILE_PATH)

print(f"Dataset loaded: {raw_data.shape[0]} rows x {raw_data.shape[1]} columns")
print(f"Columns: {list(raw_data.columns)}")
```

```python
# The first 5 rows of the dataset
raw_data.head(5)
```

```python
# Printing the datatype for only gender right now
print("Data type for the column Gender is", raw_data['Gender'].dtypes)
print("The unique values in Gender are:" + str(raw_data['Gender'].unique()))
print("The total amount of unique values are:", len(raw_data['Gender'].unique()))
```

```python
# Counting how many of each unique value there is in the dataset

print("Unique Gender values and their counts:")
print(raw_data['Gender'].value_counts())
```

```python
gender_mapping = {'female': 0, '0':0, 'male':1, '1':1}

# Stripping of the data and making them lower case and then assigning the new data to a new column
raw_data['Gender_Clean'] = raw_data['Gender'].astype(str).str.lower().str.strip().map(gender_mapping)

print(raw_data['Gender_Clean'].value_counts())

# Checking to see if there any data that did not get mapped
print("NaNs:", raw_data['Gender_Clean'].isnull().sum())

# Dropping the original 'Gender' column
raw_data = raw_data.drop(columns=['Gender'])

# Renaming 'Gender_Clean' to 'Gender'
raw_data = raw_data.rename(columns={'Gender_Clean': 'Gender'})

# Reordering columns to place the new 'Gender' column at index 2 (its original position)
cols = raw_data.columns.tolist()

# Removing 'Gender' from its current position (it's likely at the end now)
cols.remove('Gender')

# Inserting 'Gender' at the desired index (2)
cols.insert(2, 'Gender')

# Reindexing the DataFrame with the new column order
raw_data = raw_data[cols]
print("")
raw_data.head()
```
